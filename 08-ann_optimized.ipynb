## Training on CPU vs GPU

```python
import torch
import time

# Create large matrices
size = 10000
A = torch.randn(size, size)
B = torch.randn(size, size)

# CPU computation
start = time.time()
C_cpu = A @ B  # Matrix multiplication on CPU
end = time.time()
print(f"CPU Time: {end - start:.4f} sec")

# GPU computation
A_gpu = A.to("cuda")
B_gpu = B.to("cuda")

start = time.time()
C_gpu = A_gpu @ B_gpu  # Matrix multiplication on GPU
end = time.time()
print(f"GPU Time: {end - start:.4f} sec")
```
