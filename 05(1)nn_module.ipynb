```python
import torch
import torch.nn as nn
```

```python
# create the model class
class NeuralNet(nn.Module):

  def __init__(self, input_size):

    super().__init__()

    self.network = nn.Sequential(
    nn.Linear(input_size, 3), # (batch_size, num_features)
    nn.ReLU(),

    nn.Linear(3, 1), # could name as linear2
    nn.Sigmoid()
    )

  def forward(self, input): # use sequential container for the forward pass
    output = self.network(input)

    return output
```

```python
# create dummy dataset
features = torch.rand(10,5)

# create model
model = NeuralNet(features.shape[1])

# call model for forward pass
# instead of mode.forward(features)
model(features)
```

```python
# Display weights and biases
for name, param in model.named_parameters():
    if param.requires_grad:
        print(f"Layer: {name}")
        print(f"Values:\n{param.data}\n")

# Get weights and biases of the first Linear layer
print("First Layer Weights:\n", model.network[0].weight.data)
print("First Layer Bias:\n", model.network[0].bias.data)

# Get weights and biases of the second Linear layer
print("Second Layer Weights:\n", model.network[2].weight.data)
print("Second Layer Bias:\n", model.network[2].bias.data)
```
